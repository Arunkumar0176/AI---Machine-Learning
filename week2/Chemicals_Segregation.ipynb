```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train = pd.read_csv("Logistic_X_Train.csv")
Y_train = pd.read_csv("Logistic_Y_Train.csv")
X_test = pd.read_csv("Logistic_X_Test.csv")
```

```python
X_train.head()
Y_train.head()
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
```

```python
#Data visulization
plt.scatter(X_train.iloc[:,0], X_train.iloc[:,1], c=Y_train.values.ravel())
plt.show()
```

```python
#Data Preprocessing

Y_train = Y_train.values.ravel()
```

```python
X_train.isnull().sum()
```

```python
X_train.fillna(X_train.mean(), inplace=True)
```

```python
#Train Logistic Regression Model

model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#Training Accuracy Check

pred_train = model.predict(X_train)

accuracy = accuracy_score(Y_train, pred_train)

print("Training Accuracy:", accuracy)
```

```python
test_predictions = model.predict(X_test)
```

```python
print(test_predictions[:10])
```

```python
#Save Predictions

output = pd.DataFrame(test_predictions)

output.to_csv("Logistic_Ytest.csv", index=False)
```
